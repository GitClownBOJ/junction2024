```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('faith_cooking_batch_loss.csv')
print(data.head())
```

```python
data = data[data['LOSS PERCENTAGE']>=0]
data = data[data['LOSS PERCENTAGE']<=10]
```

```python
print(data['LOSS PERCENTAGE'].std())
```

```python
mean_value = data['LOSS PERCENTAGE'].mean()
std_dev = data['LOSS PERCENTAGE'].std()
```

```python
# plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value:.2f}')
# plt.axvline(mean_value + std_dev, color='blue', linestyle='dashed', linewidth=2, label=f'+1 Std Dev: {mean_value + std_dev:.2f}')
# plt.axvline(mean_value - std_dev, color='blue', linestyle='dashed', linewidth=2, label=f'-1 Std Dev: {mean_value - std_dev:.2f}')
plt.hist(data['LOSS PERCENTAGE'], bins=100, edgecolor='black', alpha=0.7)
plt.xlabel('Loss Percentage')
plt.ylabel('Frequency')
plt.title('Loss Percentage Distribution: Faith')
plt.show()
plt.show()
```
